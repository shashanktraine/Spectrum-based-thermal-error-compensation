```python
# Import Required Libraries
import pandas as pd # for working with tabular data
import numpy as np #for numerical operations
import os #for file handling
from sklearn.svm import SVR #for training a Support Vector Regression model


# Define Input Paths & Variables
fol_path = r"D:\2.ProjectWork\1.TEX\24TEX_003\24TEX_003_P03\24TEX_003_P03_04_1"  #Specifies the folder where the data files are stored.

test_file = '24TEX_003_P03_04_240930.xlsx'  #Defines the specific test dataset file.

sel_temp =[             #A list of selected temperature sensor columns.
            "T4",
            "T5",
            "T6",
            "T7",
            "T16",
            # "T11",
            # "T14",
            # "T16"
        ]

sel_dev = ["X1"] #A list of selected deviation columns





# Load Training Data from Multiple Files
data = []
if test_file in file_detials:
    file_detials.remove(test_file)

file_detials.remove(test_file)
for file in file_detials:
    file_path = os.path.join(fol_path,file)
    df = pd.read_excel(file_path)
    data.append(df)
train_data = pd.concat(data)



#  Extract Features (Temperature) and Deviation for Training
ind_list_train = []
dep_list_train = []
columns = list(train_data.columns)
for names in columns:
    if names.startswith("T"):
        ind_list_train.append(names)
    elif names.startswith("X"):
        dep_list_train.append(names)


# Create DataFrame for all the Temperature variables for training
ind_data = []
for temp in ind_list_train:
    ind_data.append(train_data[temp])
independent = pd.concat(ind_data,axis=1)



# Create DataFrame for the selected Temperature variables for training
sel_temp_list = []
for temp in sel_temp:
    sel_temp_list.append(independent[temp])
x_train = pd.concat(sel_temp_list, axis = 1)



# Create DataFrame for all the Deviation variables for training
dep_data = []
for dev in dep_list_train:
    dep_data.append(train_data[dev])
dependent = pd.concat(dep_data,axis=1)


# Create DataFrame for the selected Deviation variables for training
sel_dev_list = []
for dev in sel_dev:
    sel_dev_list.append(dependent[dev])
y_train_sel = pd.concat(sel_dev_list, axis = 1)
y_train = y_train_sel - y_train_sel.iloc[0]




# Load and Prepare Test Data
test_file_path = os.path.join(fol_path,test_file)
test_data = pd.read_excel(test_file_path)


# Extract Features (Temperature) and Deviation for Testing
ind_list_test = []
dep_list_test = []
columns_test = test_data.columns
for names_test in columns_test:
    if names_test.startswith("T"):
        ind_list_test.append(names_test)
    elif names_test.startswith("X"):
        dep_list_test.append(names_test)

# Create DataFrame for all the Temperature variables for testing
ind_data_test = []
for temp in ind_list_test:
    ind_data_test.append(test_data[temp])
independent_test = pd.concat(ind_data_test,axis=1)


# Create DataFrame for the selected Temperature variables for testing
sel_temp_list_test = []
for temp in sel_temp:
    sel_temp_list_test.append(independent_test[temp])
x_test = pd.concat(sel_temp_list_test, axis = 1)


# Create DataFrame for all the Deviation variables for testing
dep_data_test = []
for dev in dep_list_test:
    dep_data_test.append(test_data[dev])
dependent_test = pd.concat(dep_data_test,axis=1)


# Create DataFrame for the selected Deviation variables for testing
sel_dev_list_test = []
for dev in sel_dev:
    sel_dev_list_test.append(dependent_test[dev])
y_test_sel = pd.concat(sel_dev_list_test, axis = 1)
y_test = y_test_sel - y_train_sel.iloc[0]


# Train & Apply Support Vector Regression (SVR) Model
svr = SVR(C = 100, gamma = 0.01, kernel='rbf', epsilon = 1, tol = 1) 
svr.fit(x_train,y_train.values.ravel()) 
prediction = svr.predict(x_test).reshape(len(x_test),1)
error = y_test - prediction


```
