<span style="font-size:25px;"> **Importing required libraries**




```python
import pandas as pd                  # For handling and analyzing tabular data
import numpy as np                   # For numerical computations
import matplotlib.pyplot as plt      # For visualization
import os                            # For file and directory operations
import seaborn as sns                # For enhanced visualizations
```

---
<span style="font-size:25px;"> **Project Paths and Variables Input**

```python
# Root directory containing all project-related data

## Input for folder path containing all thermal related data
mother_path = "D:\\2.ProjectWork\\1.TEX"

## Input for project /machine number
machine_number = "25TEX_006"

## Input for Phase number of the project
phase_number = "P04"

## Input for Data folder
model_folder ="04"

## Input for result folder (optional)
result_folder = "06"

## Input for result file (optional)
result_file = "SJE-10 all_data_copy"

## Feature and deviation naming conventions
fea_name = "F"
dev_name = "D"

## List of feature column names used in the model
fea_list = [
"F1",
"F2",
"F3",
"F4",
"F5"
]
```

---
<span style="font-size:25px;">**Construct File Paths**

```python

## Combine paths to form complete working directories
project_path = mother_path +"\\" + machine_number + "\\"+ machine_number + "_" + phase_number 
intial_fol_path = project_path + "\\" + machine_number + "_" + phase_number + "_" + model_folder
dest_fol_path =  project_path + "\\" + machine_number + "_" + phase_number + "_" + result_folder
result_file = dest_fol_path + "\\" + result_file + ".xlsx"

```

---
<span style="font-size:25px;">**Load and Merge Data**

```python

# Get the list of all files inside the input (model) folder
file_detials = os.listdir(intial_fol_path)

# Initialize an empty list to store DataFrames
data_train = []

# Loop through each Excel file, read it into a DataFrame, and append it
for file in file_detials:
    file_path = os.path.join(intial_fol_path,file)

    # Read Excel file into pandas DataFrame
    df = pd.read_excel(file_path)

    # Append DataFrame to the list
    data_train.append(df)

# Concatenate all DataFrames into a single dataset
all_data = pd.concat(data_train)
```

---
<span style="font-size:25px;">**Data visualization**

```python
# Pairplots of all the features in the data to understand the correlation between he variables
# sns.pairplot(data)
```

```python

##Distribution plots of deviation between work and tool
plt.figure(figsize=(10,6))
print(sns.displot(all_data["D1"]))
# print(sns.displot(data["Y-Axis"]))
# print(sns.displot(data["Z-Axis"]))
```

---
<span style="font-size:25px;">**Correlation Analysis**

```python
## Compute correlation matrix for all numerical features
all_data.corr()
```

```python
## Visualize the correlation matrix as a heatmap
plt.figure(figsize=(20,12))
sns.heatmap(all_data.corr(),cmap='viridis',annot=True)
```

---
<span style="font-size:25px;">**Exporting data**

```python
## export the result file

all_data.to_excel(result_file)
```
