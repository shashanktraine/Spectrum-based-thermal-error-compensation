```python
import pandas as pd
import numpy as np
import os
```

```python
folder_path = r"D:\2.ProjectWork\1.TEX\25TEX_006\25TEX_006_P05\New folder"
dest_folder = r"D:\2.ProjectWork\1.TEX\25TEX_006\25TEX_006_P05\New folder_2"
file_detials = os.listdir(folder_path)
for file in file_detials:
     file_path = os.path.join(folder_path,file)
     df = pd.read_excel(file_path)
     df["F1"] = df["F1"]
     df_new = df.set_index("F1")
     df_list = []
     for i in range(len(df["F1"])-1):
         a = df["F1"][i]
         b = df["F1"][i+1]
         d = np.arange(a,b,1)
         df_list.append(d)

     flat_1 = np.concatenate(df_list).tolist() 

     df_1_flat = pd.DataFrame(flat_1,columns=["F1"])
     df_1_flat_new = df_1_flat.set_index("F1")



     all_df= df_1_flat_new.join(df_new)
     all_df_new = all_df.interpolate("linear")
     all_df_new = all_df_new.reset_index()

     dest_file = os.path.join(dest_folder,file)
     all_df_new.to_excel(dest_file)
```
