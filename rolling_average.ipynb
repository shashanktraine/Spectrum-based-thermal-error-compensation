```python
import pandas as pd
import numpy as np
import os
```

```python
folder_path = r"C:\Users\shashank_p\Desktop\Spectrum Programs data"
dest_folder = r"C:\Users\shashank_p\Desktop\a"
file_detials = os.listdir(folder_path)
for file in file_detials:
     file_path = os.path.join(folder_path,file)
     df = pd.read_csv(file_path)
     df_sp_speed = df["Sp_Speed"] 

     v = 1
     df_v_list = []

     while v <= len(df_sp_speed):
          if v < 5:
               
               df_v_sub = df_sp_speed.iloc[0:v]
               df_v_mean = df_v_sub.mean()
               df_v_list.append(df_v_mean)
          else:
               df_v_sub = df_sp_speed.iloc[v-5:v]
               df_v_mean = df_v_sub.mean()
               df_v_list.append(df_v_mean)

          v += 1

     df["Rolling_5"] = pd.DataFrame(df_v_list).round(0)



     x = 1
     df_x_list = []

     while x <= len(df_sp_speed):
          if x < 10:
               
               df_x_sub = df_sp_speed.iloc[0:x]
               df_x_mean = df_x_sub.mean()
               df_x_list.append(df_x_mean)
          else:
               df_x_sub = df_sp_speed.iloc[x-10:x]
               df_x_mean = df_x_sub.mean()
               df_x_list.append(df_x_mean)

          x += 1

     df["Rolling_10"] = pd.DataFrame(df_x_list).round(0)


     xx = 1
     df_xx_list = []

     while xx <= len(df_sp_speed):
          if xx < 20:
               
               df_xx_sub = df_sp_speed.iloc[0:xx]
               df_xx_mean = df_xx_sub.mean()
               df_xx_list.append(df_xx_mean)
          else:
               df_xx_sub = df_sp_speed.iloc[xx-20:xx]
               df_xx_mean = df_xx_sub.mean()
               df_xx_list.append(df_xx_mean)

          xx += 1

     df["Rolling_20"] = pd.DataFrame(df_xx_list).round(0)

     # df_all = pd.concat([df_v,df_x,df_xx],axis=1)
     # df_xx.head(10)
     dest_file = os.path.join(dest_folder,file)
     df.to_csv(dest_file)

```
